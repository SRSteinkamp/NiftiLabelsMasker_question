# Graphtheoretical Analysis

This notebook comprises the main analysis using global and nodal graph measures.  

The analysis in general is based on a second level analysis over subjects. In general: For each subject a linear model is fitted on the different parameters for global statistics and for each node in case of the nodal analysis. The mean of the regression weights is then bootstrapped over subjects and values at an level of $\alpha$ =0.01 two tailed are considered as significant contributers, that is if the confidence does not contain 0 it is seen as contributing to the statistical models. 


## Analysis on Global level:

During the preprocessing steps a connectivity matrix with a proportional threshold of 5% was created. This threshold was calculated on the absolute connection values and therefore kept the signs of the correlations. The following global statistics were calculated: 

* global efficiency 
* mean local efficiency
* mean clustering
* modularity (using the louvain algorithm, the community structure was kept for all other analysis)
* mean functional connectivity (mean of the connectivity matrix) 


```python
import glob
import auxfunctions as aux
from sklearn.externals import joblib
import numpy as np 

data_dir = '../Data/'

nodal_stats = joblib.load(data_dir + 'FCAnalysis/nodal_stats_thresh_0.05.pkl')
global_stats = joblib.load(data_dir + 'FCAnalysis/global_stats_thresh_0.05.pkl')

file_list = glob.glob(data_dir + 'MRIreconstructed/*.mat')
```

```python
import graphhelper as gh
# Load the attended speaker
AT = aux.get_reconstructed(file_list, 'AT')

# Apply Fisher's Z transformation on the correlation values, to have them more normally distributed.

AT = np.array(gh.fisherZ(AT)).flatten()

# Vector to associate subjects:
sub_vec = np.array([np.zeros(24) + i for i in range(0,23)]).flatten()
```

### Calculate first level regressors: 
A simple linear model is fit for each of the subjects including the above mentioned metrics and a constant. Each metric is normalized before fitting the model to ignore differences in scaling etc. 
The linear model is defined as: 
$$ y = \beta * X + \epsilon $$
 
Using a vectorized approach, the betas can be estimated in the following way: 

$$ \beta = (X * X')^{-1} * (X' * y) $$ 

```python
from sklearn.linear_model import LinearRegression

# Initiate linear regression (OLS) for each 
LR = LinearRegression(normalize = True, fit_intercept = True)

# Create a X (data) matrix containing all the features (not the intercept) for the linear model
X = np.array([global_stats[i] for i in global_stats.keys()])

#Fit models: 
models = [LinearRegression(normalize = True, fit_intercept = True).fit(X[:,sub_vec == i].T, AT[sub_vec ==i]) for i in np.unique(sub_vec)]

# Coefficients: 

coef = np.array([i.coef_ for i in models])
```

### Second level analysis:

Data visualization: 

```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

plt.figure(figsize = (10,7.5))

sns.boxplot(data = coef)
sns.swarmplot(data = coef)
```

### Bootstrapping tests for each coefficient: 

The mean for each coefficient is bootstrapped. The bootstrapped means are calculated 2500 sampled from the original distribution. If the confidence interval at the alpha level does not contain 0, the coefficient is considered significant.

```python
bootstrap_coef = [[np.mean(np.random.choice(mat, mat.shape[0], replace = True)) for i in range(0,2500)] for mat in coef.T]
bootstrap_coef = np.array(bootstrap_coef)

alpha = 0.01

bootstrap_CI_up = np.percentile(bootstrap_coef, 100 - alpha/2, axis = 1)
bootstrap_CI_low= np.percentile(bootstrap_coef, alpha/2, axis = 1)

plt.figure(figsize = (10,7.5))
sns.swarmplot(data = bootstrap_coef.T)
plt.xticks(np.arange(5),  [i for i in global_stats.keys()])
```
